```julia
# Define a function to calculate the square of a number
function square(x)
    return x * x
end

# Call the function
result = square(4)
println("The square of 4 is $result")
```

```julia
# Define an array
numbers = [1, 2, 3, 4, 5]

# Access elements
println("First number: ", numbers[1])

# Modify an element
numbers[3] = 10
println("Modified array: ", numbers)
```

```julia
# Create an array of squares using list comprehension
squares = [x^2 for x in 1:10]
println("Squares from 1 to 10: ", squares)
```

```julia
# Create a dictionary
ages = Dict("Alice" => 25, "Bob" => 30, "Charlie" => 35)

# Access a value by key
println("Alice's age: ", ages["Alice"])

# Add a new key-value pair
ages["Dave"] = 40
println("Updated dictionary: ", ages)
```

```julia
using DataFrames

# Create a DataFrame
df = DataFrame(Name = ["Alice", "Bob", "Charlie"], Age = [25, 30, 35])

# Access a column
println(df.Name)

# Filter rows based on a condition
filtered_df = filter(row -> row.Age > 25, df)
println(filtered_df)
```
